```python
# 환경 설정
# pip intall vitaldb 실행 시, pandas와 버전 관련 이슈 발생
!pip install --upgrade vitaldb
!pip install pandas==2.2.3

# vitaldb 버전 확인
!pip show vitaldb

# Name: vitaldb
# Version: 1.5.1
# Summary: VitalDB Python Libray
# Home-page:
# Author:
# Author-email: VitalLab <vital@snu.ac.kr>
# License:
# Location: /usr/local/lib/python3.11/dist-packages
# Requires: numpy, pandas, requests, wfdb
# Required-by:

# 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io
import vitaldb
import time
from concurrent.futures import ThreadPoolExecutor
from scipy import stats

# vitaldb의 메서드 등 확인
dir(vitaldb)

# 한글 폰트 설정 (코랩 환경용)
import matplotlib.font_manager as fm
import platform
import os

# 현재 실행 환경의 OS 확인
if platform.system() == 'Linux':
    # Linux 환경 (주로 Colab)에서 나눔 폰트 설치 및 캐시 업데이트
    !apt-get install -y fonts-nanum > /dev/null 2>&1 # apt-get 출력 숨김
    !fc-cache -fv > /dev/null 2>&1 # fc-cache 출력 숨김

    # 설치된 나눔고딕 폰트 경로 탐색
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    if not os.path.exists(font_path): # 기본 경로에 없을 경우
        fonts = [f for f in fm.findSystemFonts() if 'Nanum' in f] # 시스템에서 Nanum 폰트 검색
        if fonts:
            font_path = fonts[0] # 찾은 첫 번째 Nanum 폰트 사용
        else:
            font_path = fm.findfont(fm.FontProperties()) # 못 찾으면 기본 시스템 폰트 사용
else: # Linux가 아닌 다른 OS 환경
    if platform.system() == 'Darwin':  # macOS
        font_path = '/Library/Fonts/AppleGothic.ttf'
    elif platform.system() == 'Windows': # Windows
        font_path = 'C:/Windows/Fonts/malgun.ttf'
    else: # 기타 OS라면 기본 시스템 폰트 사용
        font_path = fm.findfont(fm.FontProperties())

# Matplotlib에 폰트 설정 적용
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# VitalDB 로그인
vitaldb.login("id", "pw")

# --- 데이터 로딩 함수 정의 ---
def get_clinical_data():
    """
    VitalDB API를 통해 임상 정보를 다운로드하고, 필요한 컬럼만 선택하여 Pandas DataFrame 반환.
    Returns:
        pd.DataFrame: 'caseid', 'death_inhosp', 'age', 'sex', 'bmi' 컬럼을 포함하는 임상 정보 DataFrame.
                      데이터 로드 실패 시 None 반환.
    """
    url = "https://api.vitaldb.net/cases"  # 임상 정보 API URL
    try:
        # API 호출 (Accept-Encoding 헤더를 identity로 설정하여 압축 해제된 데이터 요청)
        response = requests.get(url, headers={'Accept-Encoding': 'identity'})
        response.raise_for_status()  # HTTP 오류 발생 시 예외 발생

        # 응답 내용을 UTF-8 BOM 형식으로 디코딩
        text = response.content.decode('utf-8-sig')
        # 디코딩된 텍스트를 StringIO 객체로 변환하여 CSV 파싱
        df_clinical = pd.read_csv(io.StringIO(text))

        # 분석에 필요한 주요 컬럼만 선택
        df_clinical = df_clinical[['caseid', 'death_inhosp', 'age', 'sex', 'bmi']]
        print("Clinical data loaded successfully.")
        return df_clinical
    except requests.exceptions.RequestException as e:
        print(f"Error downloading clinical data: {e}")
        return None
    except Exception as e:
        print(f"Error processing clinical data: {e}")
        return None

# --- BIS 데이터 처리 함수 정의 ---
def process_case(caseid):
    """
    주어진 caseid에 대해 BIS 데이터를 로드하고 주요 통계량 계산.
    Args:
        caseid (int): 처리할 환자 케이스 ID.
    Returns:
        dict: 'caseid', 'bis_mean', 'bis_min', 'bis_below45'를 포함하는 딕셔너리.
              오류 발생 시 None 반환.
    """
    try:
        # 해당 케이스의 BIS/BIS 트랙 데이터 로드 (VitalFile 객체 생성)
        vf = vitaldb.VitalFile(caseid, ['BIS/BIS'])
        # BIS 데이터를 5초 간격으로 다운샘플링하여 Pandas DataFrame으로 변환
        df_bis = vf.to_pandas(['BIS/BIS'], interval=5)

        # BIS 평균, 최소값, BIS < 45인 샘플 수 계산
        return {
            'caseid': caseid,
            'bis_mean': df_bis['BIS/BIS'].mean(),  # BIS 평균
            'bis_min': df_bis['BIS/BIS'].min(),    # BIS 최소값
            'bis_below45': (df_bis['BIS/BIS'] < 45).sum()  # BIS < 45인 샘플 개수
        }
    except Exception as e:
        # print(f"Error processing case {caseid}: {e}") # 개별 에러 출력 (필요시 주석 해제)
        return None # 오류 발생 시 None 반환하여 전체 프로세스 중단 방지

# --- BIS 데이터 병렬 처리 함수 정의 ---
def process_bis_data(sample_size=2000):
    """
    BIS 트랙이 있는 케이스들을 찾아 병렬로 BIS 데이터를 처리하고 통계량 집계.
    Args:
        sample_size (int): 처리할 최대 케이스 수.
    Returns:
        pd.DataFrame: 각 케이스별 BIS 통계량('caseid', 'bis_mean', 'bis_min', 'bis_below45')을 담은 DataFrame.
                      데이터 처리 실패 시 빈 DataFrame 반환.
    """
    # 'BIS/BIS' 트랙을 가지고 있는 모든 케이스 ID 검색
    caseids_with_bis = vitaldb.find_cases(['BIS/BIS'])

    # 실제 처리할 케이스 수 결정 (요청된 sample_size와 실제 BIS 데이터 보유 케이스 수 중 작은 값)
    sample_caseids = caseids_with_bis[:min(sample_size, len(caseids_with_bis))]

    if not sample_caseids:
        print("No cases found with BIS/BIS track or sample_size is 0.")
        return pd.DataFrame() # 빈 DataFrame 반환

    print(f"Processing BIS data for {len(sample_caseids)} cases (out of {len(caseids_with_bis)} found)...")
    start_time = time.time()  # 처리 시작 시간 기록

    results = [] # 각 케이스 처리 결과를 저장할 리스트
    # ThreadPoolExecutor를 사용하여 최대 10개의 스레드로 병렬 처리
    with ThreadPoolExecutor(max_workers=10) as executor:
        # 각 caseid에 대해 process_case 함수를 비동기적으로 실행
        futures = [executor.submit(process_case, caseid) for caseid in sample_caseids]
        # 모든 작업이 완료될 때까지 기다리고 결과 수집
        for i, future in enumerate(futures):
            result = future.result()
            if result: # 유효한 결과만 추가
                results.append(result)
            # 진행 상황 표시 (선택적)
            # if (i + 1) % 100 == 0:
            #     print(f"Processed {i+1}/{len(sample_caseids)} cases...")

    df_bis_stats = pd.DataFrame(results)  # 결과 리스트를 DataFrame으로 변환
    print(f"BIS data processing time: {time.time() - start_time:.2f} seconds")
    return df_bis_stats

# --- 데이터 분석 및 시각화 함수 정의 ---
def analyze_bis_data(df_clinical, df_bis_stats):
    """
    임상 데이터와 처리된 BIS 통계 데이터를 병합하고, 주요 관계를 분석 및 시각화.
    Args:
        df_clinical (pd.DataFrame): 임상 정보 DataFrame.
        df_bis_stats (pd.DataFrame): BIS 통계 정보 DataFrame.
    Returns:
        pd.DataFrame: 병합되고 파생 변수가 추가된 최종 분석용 DataFrame.
                      데이터 병합 실패 시 None 반환.
    """
    if df_clinical is None or df_bis_stats.empty:
        print("Clinical data or BIS stats data is missing. Cannot proceed with analysis.")
        return None

    # 임상 데이터와 BIS 데이터를 'caseid' 기준으로 내부 조인(inner join)
    merged_data = pd.merge(df_clinical, df_bis_stats, on='caseid', how='inner')
    print(f"Merged data rows: {len(merged_data)}")

    if merged_data.empty:
        print("No data after merging. Check caseid matching.")
        return None

    # 기본 통계량 확인 (나이, BMI, 평균 BIS)
    print("\nBasic Statistics (Age, BMI, BIS Mean):")
    print(merged_data[['age', 'bmi', 'bis_mean']].describe())

    # 파생 변수 생성: BIS < 45 지속 시간 (분 단위)
    # 'bis_below45'는 5초 간격 샘플링에서의 카운트이므로, (카운트 * 5초) / 60초 = 분
    merged_data['bis_below45_mins'] = merged_data['bis_below45'] * 5 / 60

    # 파생 변수 생성: BIS < 45 지속 시간 범주화
    # 0-30분, 31-60분, 61-120분, 121-240분, 240분 초과로 구간 나누기
    bins = [0, 30, 60, 120, 240, float('inf')]
    labels = ['0-30분', '31-60분', '61-120분', '121-240분', '240분 초과']
    # include_lowest=True 옵션으로 0 포함 (만약 0부터 시작하는 구간이 있다면)
    merged_data['bis_below45_cat'] = pd.cut(merged_data['bis_below45_mins'], bins=bins, labels=labels, right=False) # right=False로 왼쪽 경계 포함

    # 파생 변수 생성: 연령 그룹 범주화
    age_bins = [0, 40, 60, 80, float('inf')] # 최대 나이 고려하여 inf 사용
    age_labels = ['40세 미만', '40-60세', '61-80세', '80세 초과']
    merged_data['age_group'] = pd.cut(merged_data['age'], bins=age_bins, labels=age_labels, right=False)

    # --- 시각화 및 통계 검정 ---

    # 시각화 1: 연령과 BIS 평균의 관계 (산점도 및 회귀선)
    plt.figure(figsize=(10, 6))
    sns.regplot(x='age', y='bis_mean', data=merged_data, scatter_kws={'alpha':0.3}, line_kws={'color':'red'}) # 투명도 및 라인 색상 조정
    plt.title('연령과 BIS 평균의 관계', fontproperties=font_prop)
    plt.xlabel('나이', fontproperties=font_prop)
    plt.ylabel('BIS 평균', fontproperties=font_prop)
    plt.grid(True, linestyle='--', alpha=0.6) # 그리드 스타일 조정
    plt.show()

    # 시각화 2: 연령대별 BIS 평균 분포 (Box plot)
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='age_group', y='bis_mean', data=merged_data)
    plt.title('연령대별 BIS 평균 분포', fontproperties=font_prop)
    plt.xlabel('연령대', fontproperties=font_prop)
    plt.ylabel('BIS 평균', fontproperties=font_prop)
    plt.show()

    # ANOVA 분석: 연령대별 BIS 평균 차이 검정
    # 각 연령 그룹별 BIS 평균값 리스트 생성 (결측치 제거)
    age_groups_for_anova = [merged_data[merged_data['age_group'] == group]['bis_mean'].dropna() for group in age_labels]
    # 데이터가 있는 그룹만 필터링하여 ANOVA 수행
    valid_age_groups_for_anova = [group for group in age_groups_for_anova if len(group) > 1] # 각 그룹 샘플 수 2 이상

    if len(valid_age_groups_for_anova) > 1: # 최소 2개 그룹 비교 가능
        f_stat, p_value_anova = stats.f_oneway(*valid_age_groups_for_anova)
        print(f"\nANOVA (Age groups vs BIS mean): F-statistic = {f_stat:.3f}, p-value = {p_value_anova:.3f}")
        if p_value_anova < 0.05:
            print("  => 연령대별 BIS 평균에 통계적으로 유의한 차이가 있습니다.")
        else:
            print("  => 연령대별 BIS 평균에 통계적으로 유의한 차이가 없습니다.")
    else:
        print("\nANOVA (Age groups vs BIS mean): Not enough data or groups to perform ANOVA.")

    # 시각화 3: 성별에 따른 BIS 평균 분포 (Box plot)
    plt.figure(figsize=(8, 6)) # 그림 크기 조정
    sns.boxplot(x='sex', y='bis_mean', data=merged_data)
    plt.title('성별에 따른 BIS 평균 분포', fontproperties=font_prop)
    plt.xlabel('성별', fontproperties=font_prop)
    plt.ylabel('BIS 평균', fontproperties=font_prop)
    plt.show()

    # t-test (독립표본): 성별에 따른 BIS 평균 차이 검정
    male_bis = merged_data[merged_data['sex'] == 'M']['bis_mean'].dropna()
    female_bis = merged_data[merged_data['sex'] == 'F']['bis_mean'].dropna()

    if len(male_bis) > 1 and len(female_bis) > 1: # 각 그룹 샘플 수 2 이상
        t_stat, p_value_ttest = stats.ttest_ind(male_bis, female_bis, equal_var=False) # Welch's t-test (등분산 가정 X)
        print(f"\nt-test (Sex vs BIS mean): t-statistic = {t_stat:.3f}, p-value = {p_value_ttest:.3f}")
        if p_value_ttest < 0.05:
            print("  => 성별에 따라 BIS 평균에 통계적으로 유의한 차이가 있습니다.")
        else:
            print("  => 성별에 따라 BIS 평균에 통계적으로 유의한 차이가 없습니다.")
    else:
        print("\nt-test (Sex vs BIS mean): Not enough data in one or both sex groups to perform t-test.")

    # 시각화 4: BIS < 45 지속시간 범주별 환자 분포 (Count plot)
    plt.figure(figsize=(10, 6))
    sns.countplot(x='bis_below45_cat', data=merged_data, order=labels) # order 옵션으로 범주 순서 지정
    plt.title('BIS < 45 지속시간 범주별 환자 분포', fontproperties=font_prop)
    plt.xlabel('BIS < 45 지속시간 (분)', fontproperties=font_prop)
    plt.ylabel('환자 수', fontproperties=font_prop)
    plt.xticks(rotation=30, ha='right') # x축 레이블 회전 및 정렬
    plt.tight_layout() # 레이아웃 자동 조정
    plt.show()

    # 범주별 환자 수 텍스트 출력
    category_counts = merged_data['bis_below45_cat'].value_counts().reindex(labels).fillna(0) # 모든 범주 포함 및 순서대로
    print("\nBIS < 45 Mins Category Counts:")
    for category, count in category_counts.items():
        print(f"  {category}: {int(count)}명")

    # 시각화 5: BIS < 45 지속시간 범주별 환자 나이 분포 (Box plot)
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='bis_below45_cat', y='age', data=merged_data, order=labels)
    plt.title('BIS < 45 지속시간 범주별 환자 나이 분포', fontproperties=font_prop)
    plt.xlabel('BIS < 45 지속시간 (분)', fontproperties=font_prop)
    plt.ylabel('나이', fontproperties=font_prop)
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

    return merged_data

# --- 메인 실행 함수 정의 ---
def main():
    """
    전체 데이터 분석 파이프라인을 실행합니다.
    """
    print("VitalDB 데이터 분석을 시작합니다.")

    # 1. 임상 데이터 로드
    df_clinical = get_clinical_data()
    if df_clinical is None:
        print("임상 데이터 로드 실패. 분석을 중단합니다.")
        return None
    print(f"임상 데이터 로드 완료: {len(df_clinical)}행")

    # 2. BIS 데이터 처리 (샘플 크기를 조절하여 테스트 가능)
    df_bis_stats = process_bis_data(sample_size=2000)
    if df_bis_stats.empty:
        print("BIS 데이터 처리 결과 없음. 분석을 중단합니다.")
        return None

    # 3. 데이터 분석 및 시각화
    final_merged_data = analyze_bis_data(df_clinical, df_bis_stats)

    if final_merged_data is not None:
        print("\n분석 완료! 최종 병합된 데이터 일부:")
        print(final_merged_data.head())
    else:
        print("\n분석 중 오류 발생 또는 데이터 부족으로 최종 데이터 생성 실패.")

    return final_merged_data

# --- 스크립트 실행 ---
if __name__ == "__main__":
    # 메인 함수 실행하여 분석 수행 및 결과 저장
    merged_data_result = main()
```
